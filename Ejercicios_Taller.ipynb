### TODO: Realizar una función que resuelva la EDO cuando el discriminante es 0

```python
from sympy.interactive import printing
printing.init_printing(use_latex=True)
from sympy import *
import sympy as sp
import math 
##### Creación variable simbólica #####
t = sp.symbols('t')

i = sp.Function('i')

#####  Constantes  #####
R = float('inf') # [OHM]
C = 200*10**-3 # [F]
L = 20 # [H]

#####  Fuente de entrada  #####
Ien = 10 #[A]
Iss = Ien

alpha = 1/(2*R*C)
omega = 1/math.sqrt(L*C)
print("alpha = ",alpha, " y omega = ",omega)
if alpha > omega:
  print("Movimiento sobre amortiguado")
elif alpha == omega:
  print("Movimiento criticamente amortiguado")
else:
  print("Movimiento sub-amortiguado")
```

```python
##### Corriente del inductor  #####
diffeq = Eq( i(t).diff(t,t)+ 1 / (R*C) * i(t).diff(t) + 1 / (L*C) * i(t), Iss / (L*C) )

diffeq
```

## Obtenemos la corriente en el inductor

```python
#####  Condiciones Iniciales  #####
Vo = 0 # [V]
Io = 0 # [A]

dio = Vo/L

#####  Se resuelve la ecuación diferencial con condiciones iniciales  #####
i_t = sp.dsolve(diffeq,ics={i(0):Io, i(t).diff(t).subs(t, 0): dio})
i_t
```

### Si se calcula $\omega$ podemos obtener el periodo de la señal

## $\omega t =0.5t=2\pi t/T\rightarrow T = 2\pi/0.5 \rightarrow T = 4\pi$

```python
plot(sympify(i_t.args[1]),(t,0,30),ylim=(0,20))
```

### Evaluando en la respuesta del circuito en $t = 2\pi [s]$ obtenemos


```python
print("iL(6.2858 [ms]) = %.3f [A]" % i_t.args[1].subs(t,2*math.pi) )
```

## Obtenemos el voltaje del circuito

```python
v_t = L*i_t.args[1].diff(t)
v_t
```

## Obtenemos la corriente en el capacitor $v_{C}(t)$

```python
ic_t = C*v_t.diff(t)
ic_t
```

### Evaluando $i_{C}(t)$ en $t = 2\pi [s]$ obtenemos


```python
print("ic(6.2858 [ms]) = %.3f [V]" % ic_t.subs(t,2*math.pi) )
```

### Evaluando la tres corrientes en $t = 2\pi [s]$ se obtiene $i_{R}(t)+i_{L}(t)+i_{C}(t)=?$  
#### ¿Cuál es la respuesta?


# Ejercicio 2

```python
from sympy.interactive import printing
printing.init_printing(use_latex=True)
from sympy import *
import sympy as sp
import math 
##### Creación variable simbólica #####
t = sp.symbols('t')

v = sp.Function('v')

#####  Constantes  #####
R = 20 # [OHM]
C = 4*10**-3 # [F]
L = 10 # [H]

alpha = 1/(2*R*C)
omega = 1/math.sqrt(L*C)
print("alpha = ",alpha, " y omega = ",omega)
if alpha > omega:
  print("Movimiento sobre amortiguado")
elif alpha == omega:
  print("Movimiento criticamente amortiguado")
else:
  print("Movimiento sub-amortiguado")
```

```python
diffeq = Eq( v(t).diff(t,t)+ 1 / (R*C) * v(t).diff(t) + 1 / (L*C) * v(t), 0 )
diffeq
```

## Obtenemos el voltaje del circuito

```python
#####  Condiciones Iniciales  #####
Vo = 0 # [V]
Io = 4.5 # [A]

dvo = -(Vo+R*Io)/(R*C)

#####  Se resuelve la ecuación diferencial con las condiciones iniciales  #####
v_t = sp.dsolve(sympify(diffeq),ics={v(0):Vo, v(t).diff(t).subs(t, 0): dvo})
v_t
```

```python
plot(sympify(v_t.args[1]),(t,0,3),ylim=(-80,0),title="voltaje del circuito")
```

### Evaluando el voltaje del circuito $v(t)$ en $t = 184.828 [ms]$ obtenemos


```python
print("v(184.828 [ms]) = %.3f [V]" % v_t.args[1].subs(t,184.828*10**-3) )
```

## $-t/\tau_{1} =-10t\rightarrow \tau_{1}= 1/10 = 0.1$

## $-t/\tau_{2} =-2.5t\rightarrow \tau_{2}= 2/5 = 0.4$

## Se obtene la corriente en el capacitor  
$i_{c}(t)=Cdv(t)/dt$


```python
ict = C*(v_t.args[1].diff(t))
ict
```

### Evaluando la corriente en el capacitor $i_{c}(t)$ en t = 100 [ms] obtenemos


```python
print("ic(100 [ms]) = %.3f [A]" % ict.subs(t,100*10**-3) )
```

## Se obtene la corriente en el inductor  
$i_{L}(t)=1/L*\int_{0}^{t}v(t)dt$


```python
iLt = (1/L)*integrate(v_t.args[1],t)
iLt
```

### Evaluando la corriente en el inductor $i_{L}(t)$ en t = 100 [ms] obtenemos


```python
print("iL(100 [ms]) = %.3f [A]" % iLt.subs(t,100*10**-3) )
```

## Se obtene la corriente en el resistor  
$i_{R}(t)=v(t)/R$


```python
iRt = v_t.args[1]/R
iRt
```

### Evaluando la corriente en el resistor $i_{R}(t)$ en t = 100 [ms] obtenemos


```python
print("iR(100 [ms]) = %.3f [A]" % iRt.subs(t,100*10**-3) )
```

### Evaluando la tres corrientes en t = 100 [ms] se obtiene $i_{R}(t)+i_{L}(t)+i_{C}(t)=?$  
#### ¿Cuál es la respuesta?


# Ejercicio 3 _**RLC serie**_

```python
from sympy.interactive import printing
printing.init_printing(use_latex=True)
from sympy import *
import sympy as sp
import math 
##### Creación variable simbólica #####
t = sp.symbols('t')

v = sp.Function('v')

#####  Constantes  #####
R = 5 # [OHM]
C = 250*10**-3 # [F]
L = 1 # [H]

#####  Fuente de entrada  #####
Ven = 24 #[V]
Vss = Ven

alpha = R/(2*L)
omega = 1/math.sqrt(L*C)
print("alpha = ",alpha, " y omega = ",omega)
if alpha > omega:
  print("Movimiento sobre amortiguado")
elif alpha == omega:
  print("Movimiento criticamente amortiguado")
else:
  print("Movimiento sub-amortiguado")
```

```python
##### Voltaje en el capacitor  #####
diffeq = Eq( v(t).diff(t,t)+ (R / L) * v(t).diff(t) + 1 / (L*C) * v(t), Vss / (L*C) )
diffeq
```

## Obtenemos el voltaje en el capactor $v_{C}(t)$

```python
#####  Condiciones Iniciales  #####
Vo = 4 # [V]
Io = 4 # [A]
# 
dvo = Io/C

#####  Se resuelve la ecuación diferencial con las condiciones iniciales  #####
vC_t = sp.dsolve(diffeq,ics={v(0):Vo, v(t).diff(t).subs(t, 0): dvo}).rhs
vC_t
```

```python
plot(sympify(vC_t),xlim=(0,5),ylim=(0,30),ylabel="$v_{C}(t)$",xlabel="[t]",title="voltaje en el capacitor")
```

### Evaluando el voltaje en el capacitor $v_{C}(t)$ en $t = 500 [ms]$ obtenemos

```python
print("vc(500 [ms]) = %.3f [V]" % vC_t.subs(t,500*10**-3) )
```

## Se obtiene la corriente del circuito
$i(t)=C*dv_{C}(t)/dt$

```python
i_t = C*(vC_t.diff(t))
i_t
```

### Evaluando la corriente del circuito $i(t)$ en $t = 500 [ms]$ obtenemos

```python
print("i(500 [ms]) = %.3f [A]" % i_t.subs(t,500*10**-3) )
```

### Evaluando el voltaje en el resistor $v_{R}(t)$ en $t = 500 [ms]$ obtenemos

```python
vR_t = R*i_t
vR_t
```

```python
print("vR(500 [ms]) = %.3f [V]" % vR_t.subs(t,500*10**-3) )
```

### Hallando el voltaje en el inductor $v_{L}(t)$ y evaluando en $t = 500 [ms]$ obtenemos

```python
vL1_t = L*i_t.diff(t)
vL1_t
```

```python
print("vL(500 [ms]) = %.3f [V]" % vL1_t.subs(t,500*10**-3) )
```

### Hallando el voltaje en el inductor por un método alternativo

```python
vL2_t = Vss - vC_t - vR_t
vL2_t
```

### Evaluando la malla completa en $t=500[ms]$ se obtiene $v_{R}(t)+v_{L}(t)+v_{C}(t)=?$  
#### ¿Cuál es la respuesta?
